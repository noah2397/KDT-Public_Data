```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import numpy as np

df=pd.read_csv("gender.csv")
daegu=list(df.iloc[43:52,0].values)
plt.figure(figsize=(9, 9))
for j,i in enumerate(daegu) :
    data=df[df.iloc[:,0]==i].iloc[:,[104,207]]
    x=int(data[" "].values)
    y=int(data[" "].values)
    plt.subplot2grid((3, 3), (int(j/3), int(j%3)), colspan=1)
    plt.pie([x,y], labels=['', ''], autopct='%1.1f%%', startangle=90)
    plt.title(i)

plt.suptitle("    ", size=20, weight="bold")
plt.tight_layout()
plt.show()
```
