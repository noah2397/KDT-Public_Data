```python
import pandas as pd


df=pd.read_csv("행정구역시군구별성별인구수.csv")
df.head(5)
```

```python
df[df["행정구역(시군구)별"].isin(['강원도', '경기도', '경상남도', '경상북도' ,'인천광역시', '대구광역시', '전라남도', '대전광역시' ,'광주광역시',
 '제주특별자치도' ,'부산광역시' ,'전라북도', '충청북도' ,'서울특별시', '충청남도', '울산광역시'])]
```

```python
df.drop(axis=0, index=[0], inplace=True)
```

```python
df["행정구역(시군구)별2"]=df["행정구역(시군구)별"]
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
local=['전국','강원도', '경기도', '경상남도', '경상북도' ,'인천광역시', '대구광역시', '전라남도', '대전광역시' ,'광주광역시',
 '제주특별자치도' ,'부산광역시' ,'전라북도', '충청북도' ,'서울특별시', '충청남도', '울산광역시']
df["행정구역(시군구)별"][0] in local
```

```python
current_local="전국"
for i in range(df.shape[0]):
    if df["행정구역(시군구)별"][i] in local :
        current_local=df["행정구역(시군구)별"][i]
        df["행정구역(시군구)별2"][i]="소계"
    else :
        df["행정구역(시군구)별2"][i]=df["행정구역(시군구)별"][i]
        df["행정구역(시군구)별"][i]=current_local
```

```python
df=df.reindex(columns=["행정구역(시군구)별", "행정구역(시군구)별2","2023.12"])
df
```

```python
df.rename(columns={"2023.12":"총인구수(명)","행정구역(시군구)별":"행정구역(시도)별1"}, inplace=True)
df.to_excel("행정구역시군구별성별인구수.xlsx", index=False)
```
